## 4. Collection Method
*   **Methodology:** Automated Web Scraping.
*   **Target Population:** Residential flats listed for sale on the MagicBricks platform within the Surat region at the time of scraping.
*   **Attributes Collected:** The scraper extracted structured fields (Price, Area, Location) and likely unstructured text (Descriptions) from individual listing pages.

## 5. Potential Limitations & Biases (Critical Thinking)
*Since this is a scraped dataset, we acknowledge the following limitations:*
1.  **Selection Bias:** The data only represents properties listed *online* on a specific platform (MagicBricks). It excludes offline transactions or properties listed on competitors' sites, potentially skewing towards a clearer, more tech-savvy demographic.
2.  **Asking Price vs. Selling Price:** The `Price` column represents the *asking price* (listing price), which is often higher than the actual *transaction price* (sold price).
3.  **Self-Reporting Errors:** Data is entered by agents or owners. Subjective fields like "Furnishing Status" or textual descriptions may contain exaggerations or inconsistencies.
4.  **Temporal Bias:** As the data is from ~2023 and not updated, it does not reflect current market inflation or recent infrastructure changes in Surat.


## 6. Motivation & Research Direction
### Why this dataset?
Our group consists of 3 members interested in **Economic Data Analysis** and **Data Engineering**. We initially considered datasets from Vietnam (HCMC, Hanoi) or China, but chose this Surat dataset for two main reasons:
1.  **Complexity ("Uncleaned"):** The dataset is intentionally "messy" (units in 'Cr', 'Lac', mixed text in numerical columns), providing an excellent opportunity to demonstrate advanced **Data Cleaning** and **Regex** skills as required by the course.
2.  **Scale:** India's housing market shares many similarities with Vietnam (rapid urbanization, density), making the insights potentially relatable.


### Potential Research Questions
This dataset enables us to investigate:
*   **Price Determinants:** Which features (Area, Balconies, Furnishing) correlate most strongly with price?
*   **Geospatial Analysis:** Which localities in Surat command the highest "Price per Sq.ft"?
*   **Predictive Modeling:** Can we build a Regression model to predict listing prices based on physical attributes with high accuracy?
