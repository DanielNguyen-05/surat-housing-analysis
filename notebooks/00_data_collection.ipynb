## 1. Data Subject

This dataset contains real-estate listings for residential flats in **an Indian city (Surat)** - describing characteristics of properties available for sale. 

It includes structured data: location, property features, sizes, prices, and possibly textual descriptions.

This reflects broader social and economic patterns such as urban housing demand, pricing trends, and infrastructure effects on property value. 

## 2. Data Source

- Platform: Kaggle
- URL: [Flats Uncleaned Dataset](https://www.kaggle.com/datasets/kunjadiyarohit/flats-uncleaned-dataset)
- Author: The dataset was uploaded by the Kaggle user Kunjadiya Rohit (as listed on the Kaggle dataset page)
- Publication: The dataset is published in Kaggle 2 years ago (2023). However, it has not been updated since the published date to now

## 3.2 Collection Method

The dataset was collected via web scraping from the real estate marketplace MagicBricks. Each listing was extracted automatically, including property details such as location, size, price, and amenities.

The target population is residential flats available for sale on MagicBricks at the time of scraping.

## 4. Purposes & Limitations

The topic of residential real estate—including flats, apartments, and homes—remains highly relevant, particularly in densely populated urban areas, where housing demand, availability, and pricing dynamics significantly impact economic and social outcomes. Understanding these patterns is crucial for urban planning, policy-making, and market analysis.

At first, we decide to find datasets that relates to our surrounding regions such as: Ho Chi Minh city, Ha Noi city (within Viet Nam); or even Shanghai, Beijing (within China); but then we find the dataset in Surat, India-which is country with the largest population. 

Moreover, the dataset is uncleaned and crawling from real-world platform presenting an opportunity for us to apply rigorous data preprocessing, cleaning, and exploratory analysis techniques.

This dataset enables the investigation of several research questions, including:
- Determinants of property pricing: Examining which features most strongly influence the market value of residential flats.
- Predictive modeling of housing prices: Developing regression models to estimate property values based on listing characteristics.
- Spatial and feature-based analysis: Assessing how location, size, and amenities impact price per square foot.
- Classification of properties by market segment: Categorizing flats according to price ranges or demand levels.
